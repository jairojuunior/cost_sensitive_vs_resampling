```python
import pandas as pd
import numpy as np
import re
from matplotlib import pyplot as plt
from matplotlib import ticker as mtick
import seaborn as sns
from glob import glob
import ast
```

```python
def import_all_experiments(path: str, ignore:list()=[])->pd.DataFrame:
    """
    Import and stack all files within a path
    """
    exp_files = glob(path)
    if len(ignore)>0:
        exp_files = list(set(exp_files) - set(ignore))
    df_parts = []
    for f in exp_files:
        df_parts.append(pd.read_csv(f))
        
    df = pd.concat(df_parts, axis=0)
    df.drop(columns='Unnamed: 0', inplace=True)
    return df

def get_percentage_of_sample_types_datasets(path: str)->pd.DataFrame:
    #Import all synthetic datasets
    datasets = glob(path)
    regex_nums = re.compile(r"\d+")
    metadata = []
    for dataset in datasets:
        df = pd.read_csv(dataset)
        qty_minority = df[df['type']!='majority'].shape[0]
        pct_borderline = round(df[df['type']=='borderline'].shape[0]/qty_minority, 4)
        pct_noise = round(df[df['type']=='noise'].shape[0]/qty_minority, 4)
        pct_safe = round(df[df['type']=='safe'].shape[0]/qty_minority, 4)
        
        dataset_params = [int(x) for x in regex_nums.findall(dataset)]
        header = {'N': dataset_params[0],
                'v': dataset_params[1],
                'l': dataset_params[2],
                'd': dataset_params[3],
                'r': dataset_params[4]/100,
                'e': dataset_params[5]/100,
                'borderline': pct_borderline,
                'noise': pct_noise,
                'safe': pct_safe}
        metadata.append(header)
    return pd.DataFrame(metadata)

def parse_metadata(df: pd.DataFrame, col)->pd.DataFrame:
    """
    Parse a metadata dict as pandas dataframe columns. 
    """
    df[col+'_dict'] = df[col].apply(lambda x: ast.literal_eval(x))
    metadata_keys = df[col+'_dict'].iloc[0].keys()
    for k in metadata_keys:
        df[k] = df[col+'_dict'].apply(lambda x: x[k])
    df.drop(columns=col+'_dict', inplace=True)
    return df

def identify_learning_algorithm_by_hyperparameters(hyperparameters: pd.Series)->pd.Series:
    algo_hyperparameters ={str(["min_impurity_decrease", "max_features", "criterion", "random_state"]): 'Árvore de Decisão',
                           str(["solver", "C", "penalty", "fit_intercept", "dual", "random_state"]): "Regressão Logística",
                           str(["n_neighbors", "weights"]): "kNN",
                           str(["n_estimators", "min_impurity_decrease", "max_features", "criterion", "random_state"]): 'Random Forest',
                           str(["n_estimators", "learning_rate", "random_state"]): "AdaBoost"}
    hyp_keys = hyperparameters.apply(lambda x: str(list(ast.literal_eval(x).keys())))
    algos = hyp_keys.replace(algo_hyperparameters)
    return algos
    

def identify_learning_algorithm(df: pd.DataFrame, col_learning_algorithm: str, 
                                col_hyperparameters: str)->pd.DataFrame:
    df[col_learning_algorithm] = identify_learning_algorithm_by_hyperparameters(df[col_hyperparameters])
    return df

def select_best_imbalance_learning_hyperparameters_result(df: pd.DataFrame, col_imbalance_approach: str, col_learning_algorithm: str,
                                                          col_imbalance_hyperparameters: str, col_metadata: str, 
                                                          col_metric: str)-> pd.DataFrame:
    """
    Select result with best col_metric for each subset of imbalance approach and 
    learning algorithm combined. This is done this way so we can reproduce what
    grid search optimization over the combination of imbalance approach and 
    learning algorithm.
    """
    df_sorted = df.sort_values(by=[col_metadata, col_learning_algorithm, col_imbalance_approach, col_metric], ascending=False)
    df_sorted['rank'] = df_sorted.groupby([col_metadata, 
                                           col_learning_algorithm, 
                                           col_imbalance_approach])[col_metric].rank(method='first',ascending=False)
    df_sorted = df_sorted[df_sorted['rank']==1] #Get best
    df_sorted.drop(columns='rank', inplace=True)
    return df_sorted

def create_bin(s: pd.Series, bin_size: float, start=None, end=None)-> tuple:
    if start==None:
        start=round(s.min(),2)
    if end==None:
        end=round(s.max(),2)
    
    bins = [round(x,2) for x in np.arange(start, end, bin_size)]
    
    if start>s.min():
        bins_2 = [round(s.min(),2)]
        bins_2.extend(bins)
        bins = bins_2
    if end<s.max():
        bins_2 = [round(s.max(),2)]
        bins.extend(bins_2)
    return bins

def bin_var(s: pd.Series, bin_size: float, start=None, end=None)->pd.Series:
    bins = create_bin(s, bin_size, start, end)
    binned = pd.cut(s, bins)
    return binned
    

def create_palettes(df: pd.DataFrame, cols_to_color: list)->dict:
    """
    [Aesthetic funcion] Create uniform pallete depending on the variable to plot, 
    keeping color consistency of plots.
    Color are optimized for color-blinded individuals based on Wong's paper 
    (https://www.nature.com/articles/nmeth.1618)
    """
    
    wong_colors = [(round(x[0]/255,2), round(x[1]/255,2), round(x[2]/255,2), 0.0) for x in [(0,0,0), (230,159,0), 
                                                                                             (86,180,233), (0,158,115), 
                                                                                             (240,228,66), (0,114,178), 
                                                                                             (213,94,0), (204,121,167)]]
    palette = {}
    for col in cols_to_color:
        palette[col] = {}
        col_vals = df[col].unique()
        for i in range(len(col_vals)):
            palette[col][col_vals[i]] = wong_colors[i]
    return palette

def make_2pointplot(df, x, y, hues, x_label, y_label, filename,
                    x_tick_label_prefix=None,  y_format=None, 
                    palettes=None, plot_size=(12,5)):
    
    fig, axs = plt.subplots(1,2,sharey=True, figsize=plot_size)

    axs[0] = make_pointplot(df, x, y, hues[0], x_label, y_label, axs[0],
                           x_tick_label_prefix,  y_format, palettes)
    axs[1] = make_pointplot(df, x, y, hues[1], x_label, 'disable', axs[1],
                           x_tick_label_prefix,  y_format, palettes)

    axs[0].set_title("(a)", fontdict={'fontsize': 13})
    axs[1].set_title("(b)", fontdict={'fontsize': 13})

    plt.tight_layout()
    plt.savefig(filename)
    plt.show()
    

def make_pointplot(df, x, y, hue, x_label, y_label, ax,
                    x_tick_label_prefix=None,  y_format=None, 
                    palettes=None):
    if palettes!=None:
        palette = palettes[hue]
    else:
        palette = None

    n_lines = df[hue].nunique()
    sns.pointplot(data=df, x=x, y=y, hue=hue, palette=palette, 
                  ax=ax, capsize=.1, errwidth=1.25, scale=0.6)

    ax.set_xlabel(x_label, fontsize=14)
    if y_label=='disable':
        ax.get_yaxis().set_visible(False)
    else:
        ax.set_ylabel(y_label, fontsize=14)
        if y_format != None:
            ax.set_yticklabels([y_format.format(x) for x in ax.get_yticks()], fontsize=13)

    ax.legend(title="", fontsize=13)

    if x_tick_label_prefix != None:
        ax.set_xticklabels([x_tick_label_prefix+x.get_text() for x in ax.get_xticklabels()], fontsize=13)
    
    return ax

experiments_all = import_all_experiments("../experiments_logs/*", ignore=['../experiments_logs/cost_algos'])
experiments_wout_thresholding = experiments_all[experiments_all['imbalance_approach']!="Thresholding"]
assert "Thresholding" not in experiments_wout_thresholding['imbalance_approach'].unique()
experiments_thresholding = import_all_experiments("../experiments_logs/synthetic_thresholding*.csv")
experiments_df = pd.concat([experiments_wout_thresholding, experiments_thresholding], axis=0)

experiments_df = parse_metadata(experiments_df, 'metadata')

#Formatting names to display
experiments_df = identify_learning_algorithm(experiments_df, 'learning_algorithm', 'learning_algorithm_hyperparams')
experiments_df['imbalance_approach'].fillna('Nenhum', inplace=True)

#Merge experiment results with percentage of sample types on datasets
datasets_sample_types = get_percentage_of_sample_types_datasets('../output/*.csv')
experiments_df = experiments_df.merge(datasets_sample_types, left_on=['N', 'v','l', 'd', 'r', 'e'], 
                                      right_on=['N', 'v', 'l', 'd', 'r', 'e'], how='left', 
                                      validate='many_to_one')

#Bin variables to facilitate analysis
#Bins are based on analysis of distribution of variables
experiments_df['borderline_bin'] = bin_var(experiments_df['borderline'], 0.05, 0, 0.3) 
experiments_df['noise_bin'] = bin_var(experiments_df['noise'], 0.05, 0, 0.3) 
experiments_df['safe_bin'] = bin_var(experiments_df['safe'], 0.2, 0, 1.2) 

#REMOVE Logistic Regression
experiments_df = experiments_df[experiments_df['learning_algorithm']!='Regressão Logística']

#Select best imbalance approach hyperparameters
experiments_df_best = select_best_imbalance_learning_hyperparameters_result(experiments_df, 'imbalance_approach', 
                                                                            'learning_algorithm', 
                                                                            'learning_algorithm_hyperparams',
                                                                            'metadata',
                                                                            'test_balanced_accuracy')


#Plot configuration
sns.set_style("ticks")  
plot_size=(8,5)
palettes = create_palettes(experiments_df_best, ['learning_algorithm', 'imbalance_approach'])
```

```python
pivot_d_l_bacc_per_imb = experiments_df_best.pivot_table(index=['d','l'], 
                                                         columns = ['imbalance_approach'], 
                                                         values='test_balanced_accuracy', 
                                                         aggfunc='mean').reset_index()

pivot_d_l_bacc_per_algo = experiments_df_best.pivot_table(index=['d','l'], 
                                                          columns = ['learning_algorithm'], 
                                                          values='test_balanced_accuracy', 
                                                          aggfunc='mean').reset_index()

pivot_d_l_bacc = pivot_d_l_bacc_per_algo.merge(pivot_d_l_bacc_per_imb, 
                                                left_on=['d', 'l'], 
                                                right_on=['d', 'l'])
pivot_d_l_bacc['l'] = '1:'+pivot_d_l_bacc['l'].astype(str)
pivot_d_l_bacc['d'] = pivot_d_l_bacc['d'].astype(str)
pivot_d_l_bacc[pivot_d_l_bacc.columns[2::]] = pivot_d_l_bacc[pivot_d_l_bacc.columns[2::]].apply(lambda x: round(x,2))
pivot_d_l_bacc.set_index(['d', 'l'], inplace=True)
```
